```python
def upsample_data(X_train, y_train, p):

            data_Y = pd.DataFrame(y_train)
            data_XY = X_train.join(data_Y)
            df = data_Y.drop_duplicates(subset=[p])
            classes_unique = df.values
            XY_new = pd.DataFrame({'A' : []})
            df1 = data_Y.loc[:,~data_Y.columns.duplicated()]
            max_len = max(df1[p].value_counts())
            
            
            for i in classes_unique:
                Extracted_X = data_XY[data_XY.loc[:,p].values == i]
                length = len(Extracted_X)
                if length == max_len:
                    Extracted_X_max = pd.DataFrame(Extracted_X)
                    if (XY_new.empty):
                        XY_new = Extracted_X_max.copy()
                        print(i)
                    else:
                        XY_new = pd.concat([XY_new, Extracted_X_max]) #length_XY = len(Extracted_X_max)
                else:
                    Extracted_X_upsampled = resample(Extracted_X, 
                                                   replace=True,     # sample with replacement
                                                   n_samples=max_len,    # to match majority class
                                                   random_state=123) # reproducible results
                    if (XY_new.empty):
                        XY_new = Extracted_X_upsampled.copy()
                    else:
                        XY_new = pd.concat([XY_new, Extracted_X_upsampled])
            return XY_new
```

```python
class Trainer:
    
    def __init__(self, df):
        
        self.X = df[[0,1,2,3,4,5,6,7]]
        self.data = df.copy() 
        
    def upsample_data(self, X_train, y_train, p):
            #print("X_train")
            #print(X_train)
            #print("y_train")
            #print(y_train)
            data_Y = pd.DataFrame(y_train)
            data_XY = X_train.join(data_Y)
            df = data_Y.drop_duplicates(subset=[p])
            classes_unique = df.values
            XY_new = pd.DataFrame({'A' : []})
            df1 = data_Y.loc[:,~data_Y.columns.duplicated()]
            max_len = max(df1[p].value_counts())
#             classes_unique = list(data_Y.unique())
#             XY_new = pd.DataFrame({'A' : []})
#             max_len = max(y_train.value_counts())

# suppose actual
# 3 :5
# 4 : 22
# 2 : 10
# 1 : 1
            for i in classes_unique:
                Extracted_X = data_XY[data_XY.loc[:,p].values == i]
                length = len(Extracted_X)
                if length == max_len:
                    Extracted_X_max = pd.DataFrame(Extracted_X)
                    if (XY_new.empty):
                        XY_new = Extracted_X_max.copy()
                        print(i)
                    else:
                        XY_new = pd.concat([XY_new, Extracted_X_max]) #length_XY = len(Extracted_X_max)
                else:
                    Extracted_X_upsampled = resample(Extracted_X, 
                                                   replace=True,     # sample with replacement
                                                   n_samples=max_len,    # to match majority class
                                                   random_state=123) # reproducible results
                    if (XY_new.empty):
                        XY_new = Extracted_X_upsampled.copy()
                    else:
                        XY_new = pd.concat([XY_new, Extracted_X_upsampled])
            return XY_new

    def run(self, classifiers,params, names, prediction_label = None, method = "classify", split_size = 0.20, k_folds = 5,digitize = False, upsample = False):
        
        results = []
        #"select feature type for y label"
        assert prediction_label is not None and prediction_label in self.data.columns

        Y = self.data[[prediction_label]]

        if  prediction_label == "Overall" and digitize: #for overall scores
            Y = np.digitize(np.array(Y),np.array([5.0, 10.0, 15.0, 20.0]))
        
        X_train, X_test, y_train, y_test = train_test_split(self.X, Y, test_size=split_size, random_state= 0)
        
        #print(classes_unique)
        XY_new = pd.DataFrame({'A' : []})
        # Do upsampling if upsample = True 
        if upsample == True:
            XY_new = self.upsample_data(X_train, y_train, p)
             #random.randint(0, 500)) #70 | 30
        if(XY_new.empty):
            X_train = X_train.copy()
            y_train = y_train
        else:  
            X_train = XY_new.iloc[:, :10]
            y_train = XY_new.iloc[:, 10:]
        
        gt_y_test = y_test[[prediction_label]].values.ravel()

        for parameters, model in zip(params, classifiers):
                print(prediction_label,model)
                if method == "classify":
                    clf = GridSearchCV(model, parameters, scoring="accuracy",cv=k_folds)#make_scorer(label_ranking_loss)
                    clf.fit(X_train, y_train.values.ravel())
                    scores = cross_val_score(clf.best_estimator_,X_train, y_train, cv=k_folds)
                    res = {}
                    y_pred = clf.predict(X_test) #y_pred
                    
                    res["estimator_type"] = model.__class__.__name__
                    res["accuracy_test"] = accuracy_score(gt_y_test, y_pred)
                    target_names = ["bad","fair","good","excellent"]
                    print(classification_report(gt_y_test,y_pred))
                    
                    test = cross_val_predict(clf.best_estimator_, self.X, Y.values.ravel(), cv=k_folds)
                    r = pearsonr(test, Y.values.ravel())
#                      print(list(zip(y_pred, gt_y_test)))
#                     r = pearsonr(y_pred, gt_y_test)
                    res["correlation coefficient"] = r[0]
                    res["accuracy_cross_val"] = clf.best_score_
                    res["estimator"] = clf.best_estimator_
                    res["parameter_search_dict"] = clf.best_params_
                    
                    results.append(res)

                    
                elif method == "regression":
                    clf = GridSearchCV(model, parameters,scoring=make_scorer(mean_squared_error), cv=k_folds)
                    clf.fit(X_train, y_train.values.ravel())
#                     scores = cross_val_score(clf,X_train, y_train, cv=k_folds)
                    res = {}
                    y_pred = clf.predict(X_test) #y_pred
#                     y_pred = [round(x) for x in y_pred] #rounding to nearest integer
                    mse = mean_squared_error(y_pred,gt_y_test) #ypred vs gt y for test data
                    
                    test = cross_val_predict(clf.best_estimator_, self.X, Y.values.ravel(), cv=k_folds)
                    test = [round(x) for x in test] #rounding to nearest integer
                    r = pearsonr(test, Y.values.ravel())
#                     print(y_pred, gt_y_test)
                    r = pearsonr(y_pred, gt_y_test)
                    res["correlation coeff."] = r[0]
                    res["estimator_type"]= model.__class__.__name__
                    res["mse_test"] = mse
                    res["mse_cross_val"] = clf.best_score_
                    res["estimator"] = clf.best_estimator_
                    res["parameter_search_dict"] = clf.best_params_

                    results.append(res)
                    
                    
        return results
```
